## Usamos for

```python
#'for' iterando una lista
num = [6, 23, 89, 120]
for n in num:
    print(n)
```

```python
#'for' iterando un string
nom = "Maximiliano"
for letra in nom:
    print(letra)
```

```python
#'for' hallando la longitud de los elementos de una lista
for nombre in ['Milita', 'Luisito', 'Pablito', 'Alexandrito']:
    print('Hola {0}, la longitud de tu nombre es de: {1} caracteres.'.format(nombre,len(nombre)))
```

**Hacer un programa que itere a lo largo de una lista que contenga colores y añada (appends) todos los colores que contengan la letra r a una nueva lista.** 

```python
#lista de colores
colo = ["amarillo", "rojo", "verde", "azul", "púrpura", "blanco", "negro", "limón"]
#nueva lista
colo_con_r = []


for m in colo:
    if 'r' in m:
        colo_con_r.append(m)
print(colo_con_r)
```

**Método2: Hacer un programa que de una lista de números imprima la posición del número 9, utilizando función enumerate**

```python
#lista de números
lista = [2, 4, 5, 7, 8, 19, 3, 4]

#utilizando la función 'enumerate' que devuelva una tupla (es igual a lista, pero inmutable)
#tupla de 2 elementos de una lista (posición, contenido)
for i,m in enumerate(lista):
    if m == 9:
        break
else:
##Usamos 'for' conjuntamente con 'else'
#ponemos en negativa la posición
    i = -1
    print('No se encontró el número 9')
    
if i >=0:
    print(i)
```

**Hacer un programa que sume los elementos de una lista, utilizar funciones y for**

```python
#solicita usar for y funciones, NO SE PUEDE USAR función sum
def mysum(xs):
    """ Suma todos los números de la lista xs, y devuelve el total. """
    running_total = 0
    for x in xs:
        running_total = running_total + x
    return running_total

# Add test like these to yourteat suite ...
print(mysum([1, 2, 3, 4]))          # ==>>10
print(mysum([1.25, 2.5, 1.75]))     # ==>>5.5
print(mysum([1, -2, 3]))            # ==>>2
print(mysum([1, 2, 3, 4]))          # ==>>10
print(mysum(range(11)))             # ==>>55
print(mysum([]))                    # ==>>0
```

```python
#solicita usar for y funciones, NO SE PUEDE USAR función sum
def misuma(A):
    """ Suma todos los números de la lista A, y devuelve el total. """
    total = 0
    for B in A:
            total = total + B
    return total

# Add test like these to yourteat suite ...
print("El resultado de la suma es: "+(str(misuma([1, 2, 3, 4]))))          # ==>>10
print("El resultado de la suma es: "+(str(misuma([1.25, 2.5, 1.75]))))     # ==>>5.5
print("El resultado de la suma es: "+(str(misuma([1, -2, 3]))))            # ==>>2
print("El resultado de la suma es: "+(str(misuma([1, 2, 3, 4]))))          # ==>>10
print("El resultado de la suma es: "+(str(misuma(range(11)))))             # ==>>55
print("El resultado de la suma es: "+(str(misuma([]))))                  # ==>>0
```
