## Usamos while

**Hacer un programa que imprima la tabla del 3:**

```python
#'while' imprimiendo la Tabla del número 3
contador = 0
print('Tabla del número 3')
#Bucle while
while contador <= 10:
    print(f'3 X {contador} = {contador*3}')
    contador +=1
print("Fin de la Tabla")
```

```python
#'while' imprimiendo la Tabla del número 7
contador = 0
print('Tabla del número 7')
#Bucle while
while contador <= 10:
    print("7 X" , contador, "=" ,contador*7, )
    contador +=1
print("Fin de la Tabla")
```

**Hacer un programa que se repita una frase las veces que se desee, use funciones**

```python
#frase a repetir
frase = "Debo escribir mi propio código y comentarlo"
#contador
cont = 0

#función recibe n veces
def repetir(n,cont):
    while cont <n:
        print(frase)
        cont +=1
        
num = int(input("Ingrese en número de veces a repetir la frase:"))
print("\n")

#llamar la función
repetir(num,cont)
```

## Modificando la finalización del bucle while: break y continue
* **break** se utiliza para finalizar y salir del bucle, ***por ejemplo***, si se cumple alguna condición.
* **continue** salta al siguiente paso de la iteración, ***ignorando todas las sentencias que le siguen*** y que forman parte del bucle. 

**Uso de break:**

***Hacer un programa que valide nombre Rose:***

```python
while True:
    print("Por favor ingrese su nombre: ")
    name = input()
    if name.lower() == "rose":
        break
print("Hasta luego !!!")
```

```python
while True:
    print("Por favor ingrese su nombre: " ,end="")
    name = input()
    if name == "Rony":
        break
print("\nNombre correcto v:")
```

## Estructura iterativa while

**Hacer un programa que busque el número 2 dentro de una lista de Números, si se encuentra, que imprima el índice dentro de la lista:**

```python
valores = [5, 1, 9, 2, 7, 4]
encontrado = False
indice = 0
longitud = len(valores)
while not encontrado and indice < longitud:
    valor = valores[indice]
    if valor == 2:
         encontrado = True
    else:
        indice += 1
if encontrado:
    print(f'El número 2 ha sido encontrado en el índice {indice}')
else:
    print('El número 2 no se encuentra en la lista de valores') 
```

## Imprime la secuencia 3n+1 desde un número n , terminando cuando llegue a 1.

La "regla de cálculo" para crear la secuencia es partir de unos n dados, y generar el siguiente término de la
secuencia a partir de n , ya sea dividiendo n por la mitad, (siempre que n sea par), o bien multiplicándolo
por tres y sumando 1. La secuencia termina cuando n llega a 1.

```python
def seq3np1(n):
    while n != 1:
        print(n, end=", ")
        if n % 2 == 0:
            n = n // 2
        else:
             n = n * 3 + 1
    print(n, end=".\n")
num = int(input('Ingrese numero de inicio n : ==> '))
seq3np1(num)
```

```python
def secuencia(n):
    while n != 1:
        print(n, end=", ")
        if n % 2 == 0:
            n = n // 2
        else:
             n = n * 3 + 1
    print(n, end=".\n")
num = int(input('Ingrese numero de secuencia n : = '))
secuencia(num)
```

## Hacer un programa con función que cuenta el número de dígitos que son 0 o 5 en un entero positivo:

```python
#Creamos la función
def num_digi(n):
    count = 0
    while n > 0:
        digit = n % 10
        if digit == 0 or digit == 5:
            count = count + 1
        n = n // 10
    return count
#solicitamos numero
num = int(input('Ingrese un número que contenga los digitos "0" y "5" : ==> '))
#Asignamos la cantidad que retorna
cantidad = num_digi(num)
#imprimir resultado
print('\n\n\tLa cantidad de digitos "0" y "5" en el número ',num,'es: ==>',cantidad)
```
